```
import yahoofinance as yf
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

### Inställningar

```
#@title Date fields
ticker = 'STAR-B.ST' #@param {type:"string"}
start_date = '2020-01-01' #@param {type:"date"}
end_date = '2020-12-31' #@param {type:"date"}
```

### Hämta Data från Yahoo Finance

```
historical = yf.HistoricalPrices(ticker, start_date, end_date)
dfs = historical.to_dfs()
df = dfs['Historical Prices']

df.index = pd.to_datetime(df.index)
```

### Plotta Data

```
fig1 = plt.figure(figsize=(15,10))
plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.title(ticker + ' Share Price ' + start_date + ' to ' + end_date)
plt.plot(df[['Close']])
plt.grid(c='grey')
plt.xticks(rotation=45)
```
